```python
import cv2
import matplotlib.pyplot as plt

# image reading
img = cv2.imread("stones.jpg")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)
outlines = cv2.adaptiveThreshold(gray, 255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 9, 9)
plt.imshow(outlines)
plt.show()
```

```python
# make cartoon of your image
color = cv2.bilateralFilter(img, 9, 250, 250)
plt.imshow(color)
plt.show()
cartoon = cv2.bitwise_and(color, color, mask=outlines)
```

```python
plt.imshow(cartoon)
plt.title("The cartoon image is - ")
plt.show()
```
